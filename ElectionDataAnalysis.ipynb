```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('/content/ElectionResult.xlsx')
df.head()
```

```python
print("Null Values :\n",df.isnull().sum())
print("Total Rows : ",df.shape[0])
print("Total Columns : ",df.shape[1])
print("Unique Values :",df['Party'].nunique())
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x='Party', y='Won', data=df[:-1])
plt.xlabel('Party')
plt.ylabel('Won')
plt.title('Party Wise Results')
plt.xticks(rotation=90)
plt.show()
```

```python
# Also we don't need Leading and Total column
df = df.drop(['Leading', 'Total'], axis=1)
```

```python
df.head()
```

```python
NDA_parties = ['Bharatiya Janata Party - BJP', 'Shiv Sena - SHS','Asom Gana Parishad - AGP',
'Shiromani Akali Dal - SAD','Telugu Desam - TDP','Janata Dal (United) - JD(U)','Lok Janshakti Party(Ram Vilas) - LJPRV','Janasena Party - JnP',
               'Janata Dal (Secular) - JD(S)','Rashtriya Lok Dal - RLD','United Peoples Party, Liberal - UPPL','Apna Dal (Soneylal) - ADAL']
NDA_df = df[df['Party'].isin(NDA_parties)]
print(NDA_df)
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x='Party', y='Won', data=NDA_df)
plt.xlabel('Party')
plt.ylabel('Won')
plt.title('NDA-Performance')
plt.xticks(rotation=90)
plt.show()
```

```python
NDA_Total = NDA_df['Won'].sum()
print("Total Votes : ",NDA_Total)

NDA_df['Percentage'] = (NDA_df['Won'] / NDA_Total) * 100
NDA_df
```

```python
INDIA_df = ['Indian National Congress - INC', 'Samajwadi Party - SP','All India Trinamool Congress - AITC','Aam Aadmi Party - AAAP',
                     'Dravida Munnetra Kazhagam - DMK','Shiv Sena (Uddhav Balasaheb Thackrey) - SHSUBT','Nationalist Congress Party  Sharadchandra Pawar - NCPSP',
                     'Jharkhand Mukti Morcha - JMM','Rashtriya Janata Dal - RJD','Jammu & Kashmir National Conference - JKN','Communist Party of India (Marxist) - CPI(M)',
                     'Communist Party of India (Marxist-Leninist) (Liberation) - CPI(ML)(L)','Revolutionary Socialist Party - RSP','Marumalarchi Dravida Munnetra Kazhagam - MDMK','Rashtriya Loktantrik Party - RLTP',
                     'Kerala Congress - KEC','Viduthalai Chiruthaigal Katchi - VCK']
INDIA_df = df[df['Party'].isin(INDIA_df)]
print(INDIA_df)

INDIA_Total = INDIA_df['Won'].sum()
print("Total Votes : ",INDIA_Total)

INDIA_df['Percentage'] = (INDIA_df['Won'] / INDIA_Total) * 100
INDIA_df
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x='Party', y='Won', data=INDIA_df)
plt.xlabel('Party')
plt.ylabel('Won')
plt.title('INDIA-Performance')
plt.xticks(rotation=90)
plt.show()
```
